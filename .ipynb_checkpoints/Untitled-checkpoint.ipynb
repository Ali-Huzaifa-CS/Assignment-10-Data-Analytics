```python
sns.histplot(df['age'], kde=True)
plt.title('Distribution of Age')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
total_between = df['age'].between(17, 48).sum()
print(total_between)
```

```python
total_between_manual = df[(df['age'] >= 17) & (df['age'] <= 48)].shape[0]
print(total_between_manual)
```

```python
df['workclass'].value_counts().plot(kind='bar')
plt.title('Distribution of Workclass')
plt.xlabel('Workclass')
plt.ylabel('Count')
plt.show()
```

```python
total_bachelors_masters = df[df['education'].isin(['Bachelors', 'Masters'])].shape[0]
print(total_bachelors_masters)
```

```python
sns.boxplot(x='income', y='age', hue='gender', data=df)
plt.title('Income vs Age, Segmented by Gender')
plt.xlabel('Income')
plt.ylabel('Age')
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
df['income'] = df['income'].replace({'<=50K': 0, '>50K': 1})
print(df['income'])
```

```python
highest_salary_workclass = df.groupby('workclass')['income'].mean().idxmax()
print(f"Workclass with highest average salary (>50K): {highest_salary_workclass}")
```

```python
male_chance = df[df['gender'] == 'Male']['income'].mean()
female_chance = df[df['gender'] == 'Female']['income'].mean()

print(f"Chance for males to earn >50K: {male_chance}")
print(f"Chance for females to earn >50K: {female_chance}")
```

```python
df['workclass'] = df['workclass'].astype('category')
print(df['workclass'].dtype)
```
